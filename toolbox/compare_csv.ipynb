```python
import pandas as pd
```

```python
backup = pd.read_csv("work_orders_signals_7_27.csv")
```

```python
afterremi = pd.read_csv("work_orders_signals_afterremi.csv")
```

```python
backup = backup.dropna(subset = ["ASSET_TYPE"]).loc[backup.ASSET_TYPE == "Signal"].copy()
afterremi = afterremi.dropna(subset = ["ASSET_TYPE"]).loc[afterremi.ASSET_TYPE == "Signal"].copy()
```

```python
backup_with_sig = backup.dropna(subset = ["SIGNAL_ID"])
remi_with_sig = afterremi.dropna(subset = ["SIGNAL_ID"])
```

```python
backup_with_sig.info()
```

```python
backup_wo_sig = backup[~backup.index.isin(backup_with_sig.index)]
remi_wo_sig = afterremi[~afterremi.index.isin(remi_with_sig.index)]
```

```python
backup_wo_sig = backup_wo_sig[["ATD_WORK_ORDER_ID"]].copy()
remi_wo_sig = remi_wo_sig[["ATD_WORK_ORDER_ID"]].copy()
```

```python
backup_wo_sig.info()
```

```python
remi_wo_sig.info()
remi_wo_sig.to_csv("visual_compare.csv")
```
