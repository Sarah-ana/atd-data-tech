```python
# drop records in knack_reservation_df that has no sharepoint ID
knack_reservations_df = knack_reservations_df.dropna(subset = ["Sharepoint ID"]).copy()
```

```python
knack_reservations_df["Sharepoint ID"] = knack_reservations_df["Sharepoint ID"].astype(int)
```

```python
# change the index for both dataframes to Sharepoint id
df_reservations["Sharepoint ID index"] = df_reservations["Sharepoint ID"]
knack_reservations_df["Sharepoint ID index"] = knack_reservations_df["Sharepoint ID"]

df_reservations = df_reservations.set_index("Sharepoint ID index")
knack_reservations_df = knack_reservations_df.set_index("Sharepoint ID index")
```

```python
# assign 
sharepoint_reservations_df = df_reservations

for i in sharepoint_reservations_df.index:
    # Give the Sharepoint rsvp Table Banner OTS IDs that match the Reservations Table exported from Knack
    banner_id = knack_reservations_df.loc[ knack_reservations_df["Sharepoint ID"] == sharepoint_reservations_df["Sharepoint ID"], 'Banner LPB ID']
    sharepoint_reservations_df["banner reservations LPB"] = banner_id
```

```python
code_name_dict = dict()
for i, pair in enumerate(code_name):
    code_name_dict[pair["Code"]] = pair["Name"]
    
```

```python
code_name_dict
```

```python
for i in df_reservations.index:
    locations_list = df_reservations["DistrictName_import"][i].split(";")
    if len(locations_list) == 1:
        # get rid district code from district name
        
        district_name_clean = df_reservations["DistrictName_import"][i].split(" - ")[0]
        
        if district_name_clean in code_name_dict:
            district_name_clean = code_name_dict[district_name_clean]
            
        
        data = [{
            'Sharepoint_ID': df_reservations['Sharepoint ID'][i],
            'Status LPB': df_reservations['Banner Status'][i],
            'Permit Number': df_reservations['Permit Number'][i],
            'banners_organization': df_reservations["Organization Name"][i],
            'Organization Name': df_reservations['Organization Name'][i],
            'District': district_name_clean,
            'LPB Install Date': df_reservations['Installation Date'][i],
            'LBP Remove Date': df_reservations['Removal Date']
            
        }]
        
        
        reso_locations_df = reso_locations_df.append(data, ignore_index=True)
        
    else:
        for location in locations_list:
            
            district_name_clean = df_reservations["DistrictName_import"][i].split(" - ")[0]
        
            if district_name_clean in code_name_dict:
                district_name_clean = code_name_dict[district_name_clean]

            data = [{
                'Sharepoint_ID': df_reservations['Sharepoint ID'][i],
                'Status LPB': df_reservations['Banner Status'][i],
                'Permit Number': df_reservations['Permit Number'][i],
                'banners_organization': df_reservations["Organization Name"][i],
                'Organization Name': df_reservations['Organization Name'][i],
                'District': district_name_clean,
                'LPB Install Date': df_reservations['Installation Date'][i],
                'LBP Remove Date': df_reservations['Removal Date']

            }]
            
            reso_locations_df = reso_locations_df.append(data, ignore_index=True)
            
        
    
        
    
```
