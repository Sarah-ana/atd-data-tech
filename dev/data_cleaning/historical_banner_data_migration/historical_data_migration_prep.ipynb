```python
def change_column_name(rsvp_hist, ots = False):
    
        # Reomove unneeded columns
    removed_columns = [
    #     'Payment Rec?',
    #     'App Received?',
#         'Co-Sponsor?',
#         'Design Rec?',
#         'Item Type',
#         'Path'
    ]
    for headers in removed_columns:
        if headers in list(rsvp_hist):
            rsvp_hist = rsvp_hist.drop(columns=removed_columns)
    
    # construct a dictionary to rename column names
    
    # dictionary with all possible column name change when input is not OTS
    dict_rename_all = {
    # Sharepoint : Knack
    'ID': 'Sharepoint ID',
    'Permit #': 'Permit Number',
    'EventName': 'Event Name', 
    'Organization': 'Organization Name', 
    'Payment Rec?' : "Payment Rec",
    'Delivered?': 'Banners Delivered',
    'Status': 'Banner Status',
    'Quantity': 'Number of Banners',
    # Count Banner Jobs
    'RSVP Date': 'CREATED_DATE',
    'Installation_Date': 'Installation Date',
    'Removal_Date': 'Removal Date',
    # Keep all 'Banner_Face' & 'OTS_Linker' 1-4 columns the same
    # 'App Received?'
    # 'Co-Sponsor?'
    # 'Design Rec?'
    'Will Pick-up?': 'Will Pick Up',
    # 'Item Type'
    # 'Path'
    # Below: LPB specific column name
    'BannerType': "Banner Type",
    'DistrictName': "DistrictName_import",
    'Total Fee' : "Total Fee",
    "WaiverApproved_Date" : "Waiver Approved Date",
    "WaiverAmount" : "Waiver Amount",
    "Payment Date": "Payment Date",
    "Comments" : "Comments",
    'Co-Sponsor?' : "Co-Sponsor?",
    'Design Rec?' : 'Design Rec?'
    }
    
    if ots == True:
        dict_rename_all['EventName'] = 'Event Name OTS'
        dict_rename_all['Status'] = 'Banner Status OTS'
        dict_rename_all['Quantity'] = 'Number of Banners OTS'
        dict_rename_all['Installation_Date'] = 'Installation Date OTS'
        dict_rename_all['Removal_Date'] = 'Removal Date OTS'
    
    # construct rename dictionary for current dataframe
    dict_rename = {}
    
    for column_header in list(rsvp_hist):
        print(column_header)
        if column_header in dict_rename_all: 
            dict_rename[column_header] = dict_rename_all[column_header]
    
    rsvp_hist_col_corrected = rsvp_hist.rename(columns = dict_rename)
#     rsvp_hist_col_corrected.to_csv(Path("output_data/"))
    return rsvp_hist_col_corrected
    
```

```python
lpb_rsvp_hist_col_corrected = change_column_name(lpb_rsvp_hist)
```
