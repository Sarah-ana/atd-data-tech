```python
import pandas as pd
import numpy as np
```

```python
# import
secondary = pd.read_csv("knack_signal_secondary_8_9.csv")
all_signal = pd.read_csv("knack_singal_all_8_9.csv")
cabinet = pd.read_csv("knack_cabinet_8_9.csv")
```

```python
secondary.head(5)
```

```python
# clearning crop columns
secondary = secondary[["SIGNAL_ID", "PRIMARY_SIGNAL_ID"]].copy()
all_signal = all_signal[["SIGNAL_ID", "CABINET_ID"]].copy()
cabinet = cabinet[["CABINET_ID", "SIGNAL_ID"]]
```

```python
# cleaning change column data type
secondary["PRIMARY_SIGNAL_ID"] = secondary["PRIMARY_SIGNAL_ID"].astype(str)
secondary["SIGNAL_ID"] = secondary["SIGNAL_ID"].astype(str)
# secondary["CABINET_ID"] = secondary["CABINET_ID"].astype(str)

all_signal["SIGNAL_ID"] = all_signal["SIGNAL_ID"].astype(str)
all_signal["CABINET_ID"] = all_signal["CABINET_ID"].astype(str)

cabinet["CABINET_ID"] = cabinet["CABINET_ID"].astype(str)
cabinet["SIGNAL_ID"] = cabinet["SIGNAL_ID"].astype(str)
```

```python
# clearning delete trailing zero
secondary["PRIMARY_SIGNAL_ID"] = secondary["PRIMARY_SIGNAL_ID"].str.split('.', expand = True)[0]
secondary["SIGNAL_ID"] = secondary["SIGNAL_ID"].str.split('.', expand = True)[0]

all_signal["SIGNAL_ID"] = all_signal["SIGNAL_ID"].str.split('.', expand = True)[0]
all_signal["CABINET_ID"] = all_signal["CABINET_ID"].str.split('.', expand = True)[0]

cabinet["SIGNAL_ID"] = cabinet["SIGNAL_ID"].str.split('.', expand = True)[0]
cabinet["CABINET_ID"] = cabinet["CABINET_ID"].str.split('.', expand = True)[0]
```

```python
# processing map cabinet ID to secondary signals
secondary_cab = secondary.merge(cabinet, left_on = "PRIMARY_SIGNAL_ID", right_on = "SIGNAL_ID", how = "left").copy()
```
